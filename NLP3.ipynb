## BAG of Words


```python

import pandas as pd
```

```python
df=pd.read_csv('spam.csv',encoding='latin1')
df=df[['v1','v2']]
df.columns=['label','message']
df.head()
```

```python
## Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## BOW

from sklearn.model_selection import train_test_split
y=pd.get_dummies(df['label']
                 , drop_first=True
                 )
X_train,X_test,Y_train,Y_test=train_test_split(corpus,y,test_size=0.2,random_state=0)
from sklearn.feature_extraction.text import CountVectorizer
# max_features------> no of top features to pick based on frequency of vocabulary
cv=CountVectorizer(max_features=2500,binary=True)
X=cv.fit_transform(X_train).toarray()
```

```python
## BOW with n grams
from sklearn.feature_extraction.text import CountVectorizer
# max_features------> no of top features to pick based on frequency of vocabulary
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(1,2))
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(corpus,y,test_size=0.2,random_state=0)
X_bow=cv.fit_transform(X_train).toarray()
```

# Naive bayes works well with sparse data

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_bow,Y_train)
```

```python
X_bow_test=cv.transform(X_test).toarray()
y_pred=spam_detect_model.predict(X_bow_test)
```
